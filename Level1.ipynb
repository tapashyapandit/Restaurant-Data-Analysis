```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset into a DataFrame 

df = pd.read_csv('Dataset.csv')
print(df)
```

```python
# to check any missing data in dataset
df.isnull()
```

```python
# return top 5 rows
df.head()
```

```python
# shows the number of rows and columns
df.shape
```

```python
# column names
df.columns
```

```python
# count how many no. of times Cuisines mostly serve

cuisine_counts = df['Cuisines'].value_counts()
```

```python
print(cuisine_counts)
```

```python
# Sort cuisines in descending order

sorted_cuisines = cuisine_counts.sort_values(ascending=False)
print(sorted_cuisines)
```

```python
# top 3 cuisines
top_three_cuisines = sorted_cuisines.head(3)
print(top_three_cuisines)
```

```python
# total restaurants

total_restaurants = len(df)
print(len(df))
```

```python
# only the rows for the top three cuisines

top_cuisine_restaurants = df[df['Cuisines'].isin(top_three_cuisines.index)]
print(top_cuisine_restaurants.head(3))
```

```python
# count the number of unique restaurants

restaurants_serving_top_cuisines = top_cuisine_restaurants['Restaurant Name'].nunique()
```

```python
# percentage of restaurants that serve each of the top cuisines

percentage_restaurants_serving_top_cuisine = (restaurants_serving_top_cuisines / total_restaurants) * 100
print(percentage_restaurants_serving_top_cuisine)
```

**LEVEL 1 : TASK 2**

```python
# no. of restaurants in each city
restaurants_per_city = df.groupby('City')['Restaurant ID'].count()
print(restaurants_per_city)
```

```python
# city with the highest number of restaurants
city_with_highest_restaurants = restaurants_per_city.idxmax()
print("City with the highest number of restaurants: ", city_with_highest_restaurants)
```

```python
# the average rating for restaurants in each city
average_rating_per_city = df.groupby('City')['Aggregate rating'].mean()
print("Average rating for restaurants in each city:")
print(average_rating_per_city)
```

```python
# the city with the highest average rating
city_highest_rating = average_rating_per_city.idxmax()
print("City with the highest average rating: ", city_highest_rating)

# highest average rating of that city
highest_average_rating = average_rating_per_city.max()
print("Highest average rating:", highest_average_rating)
```

**LEVEL 1 : TASK 3** 

```python
# counts the occurrences of each price range
price_counts = df['Price range'].value_counts()
```

```python
# sort in ascending order
price_counts = price_counts.sort_index()
```

```python
# create the bar chart 
plt.bar(price_counts.index, price_counts.values)

# add labels and title
plt.xlabel('Price Range')
plt.ylabel('Number of Restaurants')
plt.title('Distribution of Price Ranges Among Restaurants')
plt.show()
```

```python
# the percentage of restaurants in each price range category
price_percentage = df['Price range'].value_counts(normalize=True) * 100
print("Percentage of restaurants in each price range category:")
print(price_percentage)
```

**LEVEL 1 : TASK 4**

```python
# the percentage of restaurants having online delivery
online_delivery_percentage = (df['Has Online delivery'].value_counts(normalize=True) * 100).round(2)

print("Percentage of restaurants that offer online delivery:")
print(online_delivery_percentage)
```

```python
#  the average ratings of restaurants with and without online delivery 
average_ratings = df.groupby('Has Online delivery')['Aggregate rating'].mean()
print("Average ratings of restaurants with and without online delivery:")
print(average_ratings)
```
